```python
from scipy import stats
import numpy as np

baseline_aov = 86.03   
aov_stddev   = 93.98  

mde          = baseline_aov * 0.05
z_alpha      = stats.norm.ppf(0.975)
z_beta       = stats.norm.ppf(0.80)
n            = 2 * (aov_stddev**2) * (z_alpha + z_beta)**2 / (mde**2)

print(f"Per variant: {int(np.ceil(n))}")
print(f"Total: {int(np.ceil(n)) * 2}")
```
